```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3-Q1
print(IQR)
```

```python
df = df[~((df < (Q1-1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
df.shape
```

```python
### Histograms
df.Open.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Number of Volume by Open Prices")
plt.ylabel('Number of Volume')
plt.xlabel('Open');
```

```python
### Heat Maps
# Finding the relations between the variables.
plt.figure(figsize=(10,10))
c= df.corr()
sns.heatmap(c,cmap="BrBG",annot=True)
c
```

```python
# Plotting a scatter plot
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(df['Close'], df['Open'])
ax.set_xlabel('Close')
ax.set_ylabel('Open')
plt.show()
```

```python
#plot the scatter plot of Open and Volume variable in data
plt.scatter(df.Open,df.Volume)
plt.show()

#plot the scatter plot of balance and age variable in data
df.plot.scatter(x="Close",y="Volume")
plt.show()
```

```python
g = sns.pairplot(df, diag_kind="kde")
g.map_lower(sns.kdeplot, levels=4, color=".2")
```

```python
df.Open.describe()
```

```python
### Correlation Matrix 
# Creating a matrix using Open, High, Low as rows and columns
df[['Open','Low','High']].corr()

#plot the correlation matrix of Open, High and Low in data dataframe.
sns.heatmap(df[['Open','Low','High']].corr(), annot=True, cmap = 'Reds')
plt.show()
```

```python
### Pivot table with Three Columns and then Heatmap.
result = pd.pivot_table(data=df, index='Date', columns='Open',values='Close')
print(result)

#create heat map of education vs marital vs response_rate
sns.heatmap(result, annot=True, cmap = 'RdYlGn', center=0.117)
plt.show()
```

```python
corr=df.corr(method='spearman')
plt.figure(figsize=(15,15))
sns.heatmap(corr,vmax=.8,linewidth=0.01,square=True,annot=True,cmap='RdBu',linecolor='black')
```
